#### Import Packages Used Throughout This Notebook

```python
import tensorflow as tf
import altair as alt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

Load data.

```python
data = datasets.load_wine()

X, y = data['data'], data['target']
features, targets = data['feature_names'], data['target_names']

df = pd.DataFrame(X, columns=features)
df = df.assign(targets=targets[y])
```

Visualize data.

```python
colors = np.array(['red', 'blue', 'green'])
```

```python
fig, axes = plt.subplots(3, 4, figsize=(20, 20))

for i, ax in enumerate(axes.flatten()):
    ax.scatter(x=X[:,i], y=X[:,-1], c=colors[y])
    ax.set_title(f'{features[i].title()} and {features[-1].title()}')
```
