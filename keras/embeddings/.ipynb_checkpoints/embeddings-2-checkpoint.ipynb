```python
from keras.layers import Input, Embedding, Dense, Reshape, merge
from keras.models import Model
from keras.utils import to_categorical
from keras.preprocessing.text import Tokenizer, one_hot
from keras.preprocessing.sequence import pad_sequences, skipgrams, make_sampling_table
from nltk.corpus import brown

import matplotlib.pyplot as plt
import numpy as np

text = 'cp03'
corpus = brown.sents(text)
```

#### Preprocess Text

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)
vocab_size = len(tokenizer.word_index) + 1
```

Encode and pad text

```python
encoded = tokenizer.texts_to_sequences(corpus)
```

```python
padded = pad_sequences(encoded, maxlen=61, padding='post')
```

```python
padded[0]
```

Create skip grams and get labels

```python
grams, labels = skipgrams(padded.flatten(), vocab_size)
```

Separate targets and context

```python
targets, contexts = zip(*grams)
```

```python
targets = np.array(targets, dtype='int')
contexts = np.array(contexts, dtype='int')
```

#### Build Model

First define layers in the model.

```python
I_target = Input((1,))
I_context = Input((1,))

E = Embedding(vocab_size, 100, input_length=1, name='embedding')
```

Feed forward some layers

```python
E_target = E(I_target)
E_context = E(I_context)
```

Reshape some layers

```python
E_target = Reshape((100, 1))(E_target)
E_context = Reshape((100, 1))(E_context)
```

Calculate similarity

```python
dot_product = merge.dot([E_target, E_context], 1, True)
dot_product = Reshape((1, ))(dot_product)
```

```python
output = Dense(1, activation='sigmoid')(dot_product)
```

---
Define model

```python
model = Model(inputs = [I_target, I_context], outputs = output)
model.compile('rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

#### Train Model

```python
model.fit([targets, contexts], labels, epochs=1000, verbose=0)
```

Save model

```python
model.save("embeddings-2-1K.model")
```

#### Get Embeddings

```python
w = E.get_weights()[0]
```

Check weight shape

```python
w.shape
```

Check similarity

```python
word_index = tokenizer.word_index

cos = lambda x, y: (x @ y) / (np.sqrt(x@x) * np.sqrt(y@y))
```

```python
w1, w2 = 'man', 'son'

v1, v2 = w[word_index[w1]], w[word_index[w2]]

print(f'Similarity is {cos(v1, v2)}')
```

#### Visual Embeddings

Import packages to reduce dimensions

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import altair as alt
import pandas as pd

alt.renderers.enable('notebook')
```

Get dictionaries from tokenizer.

```python
i2w = {b: a for a,b in tokenizer.word_index.items()}
```

---

Reduce dimensions at 2, 3, and 50 dimensions

```python
pca2, pca50 = PCA(2), PCA(50)
pca_2 = pca2.fit_transform(w)
pca_50 = pca50.fit_transform(w)
```

Generate data frames.

```python
df2 = pd.DataFrame(pca_2, columns=['d1', 'd2'])
df50 = pd.DataFrame(pca_50, columns=[f'd{_+1}' for _ in range(50)])
```

Add token labels

```python
tokens = [None] + [i2w[i] for i in range(df2.shape[0]) if i in i2w]
tokenr = [i2w[i] for i in range(df2.shape[0]) if i in i2w] + [None]
```

```python
df2 = df2.assign(tokens=tokens)
df50 = df50.assign(tokens=tokens)
```

Visualize PCA

```python
chart = alt.Chart(df2).mark_circle()
```

```python
chart.encode(x='d1:Q', y='d2:Q', tooltip='tokens').interactive()
```

TSNE

```python
tsne = TSNE(n_components=2, perplexity=vocab_size-10)
tsne3 = tsne.fit_transform(pca_50)
```

```python
df3 = pd.DataFrame(tsne3, columns=[f'd{_+1}' for _ in range(2)])
df3 = df3.assign(tokens=tokenr)
```

```python
chart = alt.Chart(df3).mark_point()
```

```python
chart.encode(x='d1', y='d2', tooltip='tokens').interactive() #| chart.encode(x='d3', y='d2', tooltip='tokens') 
```
